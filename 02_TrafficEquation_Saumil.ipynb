### Load modules and define function

```python
import numpy                       
import matplotlib.pyplot as plt                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16


L = 11.
rhom = 250.
nx = 51  
dx = L/(nx-1)
dt = .001  

x = numpy.linspace(0,L,nx)


def traffic(Vm,nt,ini):
    
    rho = numpy.ones(nx)*ini
    rho[10:20] = 50
    Vz=Vm*(1-rho/rhom)
    V=Vz.copy()

    for n in range(1, nt):  
        rhon = rho.copy() 
        Fn = Vm*rhon*(1-(rhon/250.))
        rho[1:] = rhon[1:]-dt/dx*(Fn[1:]-Fn[0:-1]) 
        V[1:]=Vm*(1-rho[1:]/rhom)
        rho[0] = ini
    
    return numpy.array([V,rho])    
```

###### Part A: *minimum* velocity at time t = 0 in meters per second

```python
A1 = traffic(80.,1,10.)
plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r'L (km)', fontsize=18)
plt.ylabel(r'V (km/hr)', fontsize=18)
plt.plot(numpy.linspace(0,L,nx), A1[0,:], 'r-', ls='--', lw=3)
plt.ylim(0,100);
min(A1[0,:])*1000/3600
```

###### Part A: average velocity at time t = 3 minutes in meters per second

```python
A2 = traffic(80.,50,10.)
plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r'L (km)', fontsize=18)
plt.ylabel(r'V (km/hr)', fontsize=18)
plt.plot(numpy.linspace(0,L,nx), A2[0,:], 'g-', ls='--', lw=3)
plt.ylim(0,100);
numpy.mean(A2[0,:])*1000/3600
```
